```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("/content/drive/MyDrive/open_one_time_covid_education_impact.csv")
df.head()
```

Data Pre-processing

```python
df['financial_situation'].value_counts()
```

```python
df.replace(['I can afford food, but nothing else', 'I cannot afford enough food for my family','I can afford food and regular expenses, but nothing else', 'I can afford food, regular expenses, and clothes, but nothing else'],
             [1, 1, 1, 1], inplace = True)
df.replace(['Prefer not to answer', 'I can comfortably afford food, clothes, and furniture, and I have savings', 'I can comfortably afford food, clothes, and furniture, but I don’t have savings', 'Not Available'],
             [0, 0, 0, 0], inplace = True)
```

```python
df['gender'].value_counts()
```

```python
# Converting Gender column into binary i.e. Male = 1 & Female = 0
df.replace(['Non-Binary', 'Male', 'Female'], [2, 1, 0], inplace = True)
```

```python
df['age'].value_counts()
```

```python
# Converting the String into Numeric Data by picking the Lower Bound
df.replace(['Under 18', '18 to 25 years old', '26 to 35 years old', '36 to 45 years old', 'Over 45 years old', 'Not Available'],
             [15, 18, 25, 35, 45, 90], inplace = True)
```

```python
df['do_children_3_and_17_yrs_receive_regular_school_meals'].value_counts()
```

```python
# Converting the String into Numeric Data by picking the Lower Bound
df.replace(['No', '1 day', '2 days', '3 days', '4 days', 'Every day'],
             [0, 1, 2, 3, 4, 5], inplace = True)
```

```python
df
```

```python
df = df[['gender','age', 'are_children_attending_face_to_face_classes', 
              'can_children_observe_deterioration_of_basic_services_of_school',	'do_children_3_and_17_yrs_receive_regular_school_meals', 
              'do_children_have_internet_connection',	'do_children_3_to_17_yrs_miss_virtual_class_due_to_lack_of_electricity',
              'does_home_shows_severe_deficit_of_electricity', 'does_home_shows_severe_deficit_of_internet', 'financial_situation']]
df
```

Splitting the Data into Train, Test & Validation

```python
dataset = df.values
dataset
```

```python
# Assigning the first 9 columns as features
X = dataset[:,0:9]
X
```

```python
# Assigning the 'financial situation' as the target variable
Y = dataset[:,9]
Y
```

```python
# Min-Max Scaler scales the dataset so that all the input features lie between 0 and 1
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
```

```python
X_scale
```

```python
# Test size will be 20% of the overall dataset
X_train, X_val_and_test, Y_train, Y_val_and_test = train_test_split(X_scale, Y, test_size=0.2)
```

```python
# Dividing the Test data into 2 equal parts: one for Test and other for Validation
X_val, X_test, Y_val, Y_test = train_test_split(X_val_and_test, Y_val_and_test, test_size=0.5)
```

```python
print(X_train.shape, X_val.shape, X_test.shape, Y_train.shape, Y_val.shape, Y_test.shape)
```

Building the Model

```python
from keras.models import Sequential
from keras.layers import Dense
```

**Architecture of the Neural Network**

1.   Hidden layer 1: 32 neurons, ReLU activation
2.   Hidden layer 2: 32 neurons, ReLU activation
3.   Output Layer: 1 neuron, Sigmoid activation

```python
# Here the 'input_shape' is set to 10, as the input layer shall have 10 features that we are inputing from X
model = Sequential([
    Dense(32, activation='relu', input_shape=(9,)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid'),
])
```

```python
# Stochastic Gradient Descent (sgd) is used the Algorithm for Optimization
# Binary-Cross Entropy is used because we are predicting whether the student belongs to High Income or Low Income
# (The loss function for outputs that take the values 1 or 0 is called binary cross entropy.)
# Accuracy metrics is used to measure the Model's Performance
model.compile(optimizer='sgd',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# 'Batch Size' gives the number of samples that will be passed through to the network at one time/cycles.
# 'Epoch' tells for how long the data will be trained or for how many cycles it will be trained.
hist = model.fit(X_train, Y_train,
          batch_size=32, epochs=25,
          validation_data=(X_val, Y_val))
```

```python
# 'Evaluate' function will give us the "Loss" & "Accuracy" of our Model
model.evaluate(X_test, Y_test)
```
